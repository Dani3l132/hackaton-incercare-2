```python
from PIL import Image, ImageDraw
import sys
import random
image_path='motor.png'

def draw_shapes(image_path):
    # Open the image
    image = Image.open(image_path)

    # Get a drawing context
    draw = ImageDraw.Draw(image)

    # Draw a rectangle
    rectangle_coordinates = (50, 50, 200, 200)  # (x1, y1, x2, y2)
    draw.rectangle(rectangle_coordinates, outline="red", width=2)

    # Draw a circle
    circle_center = (300, 150)  # (x, y)
    circle_radius = 50
    draw.ellipse((circle_center[0] - circle_radius, circle_center[1] - circle_radius,
                  circle_center[0] + circle_radius, circle_center[1] + circle_radius),
                 outline="blue", width=2)

    # Save or display the modified image
    modified_image_path = "modified_" + image_path

    image.save(modified_image_path)
    image.show()

draw_shapes(image_path)
```
